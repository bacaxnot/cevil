```python
#LIBRERIAS PROPIAS
from sections import *
from channels import *
#LIBRERIAS EXTERNAS
import matplotlib.pyplot as plt
```

```python
#CASO: Hallar Yc, Emin, Qmax y Gráficas de Q y E
%matplotlib inline
K = SquareSection(b = 2.6, y = 0.8)
#Variables
Q = 13;
Yo = 0.5;
Yf = 4;
Step = 0.001
E = 2.15;
#Desarrollo
yc = K.cDepth(Q)
print(f"Profundidad critica: {yc}")
print(f"Energía mínima: {K.specificEnergy(Q, yc)}")
print(f"Energía: {K.specificEnergy(Q)}")
print(f"Caudal máximo: {K.flowRate(E, yc)}")
print(f"Profundidades alternas: {K.alternDepths(Q, E)}")
#Graficas
K.drawSpecificEnergyCurve(Q, Yo, Yf, Step)
K.drawFlowRateCurve(E, Yo, E, Step)
```
